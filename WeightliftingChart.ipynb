## Daily Strength Processing (4/18 through today)

```python
df.Date = pd.to_datetime(df.Date)
```

```python
grp = df.groupby(['Date','Exercise']).agg({'Weight':'max', 'Reps':'sum'})
grp = grp.reset_index()
grp.index = grp.Date
grp
```

```python
activew = 'rgba(101,171,250,1.0)'
activer = 'rgba(199,92,220,1.0)'
inactivew = 'rgba(101,171,250,.5)'
inactiver = 'rgba(199,92,220,0.5)'
bg = '#2a2a2a'
marker = dict(
        symbol = 'circle',
        size = 8
    )
```

```python
####TRACES FOR EACH STRONGLIFT EXERCISE

ex1 = 'Barbell Bench Press'
benchw = go.Scatter(
    y = grp[grp.Exercise==ex1]['Weight'],
    x = grp[grp.Exercise==ex1]['Date'],
    mode = 'lines+markers',
    name = 'Weight Lifted (lbs)',
    hoverinfo='y',
    line = dict(
        color = (activew),
        shape='spline',
        width = 4,
    ),
    marker = marker,
)

benchr = go.Scatter(
    y = grp[grp.Exercise==ex1]['Reps'],
    x = grp[grp.Exercise==ex1]['Date'],
    mode = 'lines',
    name = 'Reps',
    hoverinfo='y',
    line = dict(
        color = (activer),
        shape='spline',
        smoothing=.5,
        dash='solid',
        width = 3,)
)

ex2 = 'Barbell Squat'
squatw = go.Scatter(
    y = grp[grp.Exercise==ex2]['Weight'],
    x = grp[grp.Exercise==ex2]['Date'],
    mode = 'lines',
    name = 'Weight Lifted (lbs)',
    hoverinfo='none',
    showlegend=False,
    line = dict(
        color = (inactivew),
        shape='spline',
        width = 1,
    ),
    marker = marker,
)

squatr = go.Scatter(
    y = grp[grp.Exercise==ex1]['Reps'],
    x = grp[grp.Exercise==ex1]['Date'],
    mode = 'lines',
    name = 'Reps',
    hoverinfo='none',
    showlegend=False,
    line = dict(
        color = (inactiver),
        shape='spline',
        smoothing=.5,
        dash='solid',
        width = 1,)
)

ex3 = 'Standing Barbell Military Press'
overheadw = go.Scatter(
    y = grp[grp.Exercise==ex3]['Weight'],
    x = grp[grp.Exercise==ex3]['Date'],
    mode = 'lines',
    name = 'Weight Lifted (lbs)',
    hoverinfo='none',
    showlegend=False,
    line = dict(
        color = (inactivew),
        shape='spline',
        width = 1,
    ),
    marker = marker,
)

overheadr = go.Scatter(
    y = grp[grp.Exercise==ex3]['Reps'],
    x = grp[grp.Exercise==ex3]['Date'],
    mode = 'lines',
    name = 'Reps',
    hoverinfo='none',
    showlegend=False,
    line = dict(
        color = (inactiver),
        shape='spline',
        smoothing=.5,
        dash='solid',
        width = 1,)
)
ex4 = 'Barbell Deadlifts'
deadliftw = go.Scatter(
    y = grp[grp.Exercise==ex4]['Weight'],
    x = grp[grp.Exercise==ex4]['Date'],
    mode = 'lines',
    name = 'Weight Lifted (lbs)',
    hoverinfo='none',
    showlegend=False,
    line = dict(
        color = (inactivew),
        shape='spline',
        width = 1,
    ),
    marker = marker,
)

deadliftr = go.Scatter(
    y = grp[grp.Exercise==ex4]['Reps'],
    x = grp[grp.Exercise==ex4]['Date'],
    mode = 'lines',
    name = 'Reps',
    hoverinfo='none',
    showlegend=False,
    line = dict(
        color = (inactiver),
        shape='spline',
        smoothing=.5,
        dash='solid',
        width = 1,)
)
ex5 = 'Bent Over Barbell Row'
roww = go.Scatter(
    y = grp[grp.Exercise==ex5]['Weight'],
    x = grp[grp.Exercise==ex5]['Date'],
    mode = 'lines',
    name = 'Weight Lifted (lbs)',
    hoverinfo='none',
    showlegend=False,
    line = dict(
        color = (inactivew),
        shape='spline',
        width = 1,
    ),
    marker = marker,
)

rowr = go.Scatter(
    y = grp[grp.Exercise==ex5]['Reps'],
    x = grp[grp.Exercise==ex5]['Date'],
    mode = 'lines',
    name = 'Reps',
    hoverinfo='none',
    showlegend=False,
    line = dict(
        color = (inactiver),
        shape='spline',
        smoothing=.5,
        dash='solid',
        width = 1,)
)
#### LAYOUT (Title, chart area, etc.)


layout = go.Layout(
    width=950,
    height=400,
    autosize=False,
    font = dict(
        color='white', 
        size=12, #default = 12
    ),
    title = dict(
        text = 'My Workouts',
        font = dict(
            color='white', 
            size=20,
            #family =  ''
        ),  
        xref='paper'
    ),
    paper_bgcolor = bg,
    plot_bgcolor = bg,
    margin=dict(t=40, b=30, l=30, r=10),
    yaxis=dict(
        gridcolor='#7f7f7f',
        zerolinecolor = '#7f7f7f',
        range=[0,250],
    ),
    legend = dict(
        bgcolor=bg,
        borderwidth=0,
        font = dict(
            size=15,
        ),
        x = 1,
        xanchor = 'left',
        y = 1,
        yanchor = 'top'
    ),

)

updatemenus=list([
    dict(
        buttons=list([   
            dict(
                args=[{'mode': ['lines+markers','lines','lines','lines','lines','lines','lines','lines','lines','lines'],
                       'showlegend' : [True, True, False, False, False, False,False, False,False,False],
                       'hoverinfo': ['y','y','none','none','none','none','none','none','none','none'],
                       'line.color' : [activew, activer, inactivew, inactiver,inactivew, inactiver,inactivew, inactiver,inactivew, inactiver],
                       'line.width':[4,3,1,1,1,1,1,1,1,1]}],
                label='Bench Press',
                method='restyle'
            ),
            dict(
                args=[{'mode': ['lines','lines','lines+markers','lines','lines','lines','lines','lines','lines','lines'],
                       'showlegend' : [False, False, True, True, False, False, False, False, False, False],
                       'hoverinfo': ['none','none','y','y','none','none','none','none','none','none'],
                       'line.color' : [inactivew, inactiver, activew, activer,inactivew, inactiver,inactivew, inactiver,inactivew, inactiver],
                       'line.width':[1,1,4,3,1,1,1,1,1,1]}],
                label='Barbell Squat',
                method='restyle'
            ), 
            dict(
                args=[{'mode': ['lines','lines','lines','lines','lines+markers','lines','lines','lines','lines','lines'],
                       'showlegend' : [False, False, False, False, True, True,False,False,False,False],
                       'hoverinfo': ['none','none','none','none','y','y','none','none','none','none'],
                       'line.color' : [inactivew, inactiver, inactivew, inactiver,activew, activer, inactivew, inactiver, inactivew, inactiver],
                       'line.width':[1,1,1,1,4,3,1,1,1,1]}],
                label='Military Press',
                method='restyle'
            ),
            dict(
                args=[{'mode': ['lines','lines','lines','lines','lines','lines','lines+markers','lines','lines','lines'],
                       'showlegend' : [False, False, False, False,False,False, True, True, False, False],
                       'hoverinfo': ['none','none','none','none','none','none','y','y','none','none'],
                       'line.color' : [inactivew, inactiver, inactivew, inactiver, inactivew, inactiver,activew, activer, inactivew, inactiver],
                       'line.width':[1,1,1,1,1,1,4,3,1,1]}],
                label='Deadlift',
                method='restyle'
            ),
            dict(
                args=[{'mode': ['lines','lines','lines','lines','lines','lines','lines','lines','lines+markers','lines'],
                       'showlegend' : [False, False, False, False,False,False, False, False, True, True],
                       'hoverinfo': ['none','none','none','none','none','none','none','none','y','y'],
                       'line.color' : [inactivew, inactiver, inactivew, inactiver, inactivew, inactiver, inactivew, inactiver,activew, activer],
                       'line.width':[1,1,1,1,1,1,1,1,4,3]}],
                label='Barbell Row',
                method='restyle'
            ),
        ]),
        direction = 'down',
        pad = {'l': 5, 't': 10, 'r':0},
        showactive = True,
        type = 'buttons',
        bgcolor='rgba(234,244,254,.7)',
        bordercolor = '#FFFFFF',
        borderwidth=0,
        font = dict(
            color='#065cbc',
            size=12
        ),
        x = 1,
        xanchor = 'left',
        y = .75,
        yanchor = 'top'            
    ),        

])

##
layout['updatemenus'] = updatemenus
data = [benchw, benchr, squatw, squatr, overheadw, overheadr, deadliftw, deadliftr, roww, rowr]
fig = dict(data=data, layout=layout)
py.iplot(fig, filename='workouts', config={'displayModeBar': False})
```
