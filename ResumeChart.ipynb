```python
import numpy as np
import pandas as pd

from bokeh.io import show, output_notebook
from bokeh.plotting import figure, show
from bokeh.models import Label,ColumnDataSource, HoverTool
from bokeh.core.properties import value

output_notebook()
```

```python
slots = ['Thinkful', 'vrsus', 'Daily Dose of Power','Accelerated Growth Advisors','Abbott Laboratories']


layers = ['-', #invisible bar that precedes every slot
          'Student', 'Data Analyst', 'Creator & Maintainer', 'Manager', 'Production Financial Analyst',
          'Commercial Financial Analyst','Product Cost Analyst','Corporate Financial Analyst','10K Reporting Analyst',
        'Founder & Analyst']


jobs = {'roles' : slots,
        '-'   :  [2018.6, 2017.7, 2017.55, 2016.25,2013],#invisible base for every slot
        'Student':[.7, 0, 0, 0,0],#length of each role, per slot
        'Data Analyst':   [0, 1.8, 0, 0,0],
        'Creator & Maintainer'  :  [0, 0, 2, 0,0],
        'Manager' :    [0,0,0,1.3,0],
        'Production Financial Analyst' : [0,0,0,0,.5],
        'Commercial Financial Analyst' : [0,0,0,0,.5],
        'Product Cost Analyst' : [0,0,0,0,.5],
        'Corporate Financial Analyst' : [0,0,0,0,.5],
        '10K Reporting Analyst' : [0,0,0,0,.25],
        'Founder & Analyst':    [0,0,0,0,1],
       }

       

co = ['Intensive Data Science & Machine Learning Certification',
         'SaaS Platform in the Legal Industry',
         'Educational Media Service',
         'Entrepreneurial Consulting Firm',
         'Biological Manufacturing',
         'Nutrition Sales & Marketing',
         'Singapore Supply Chain',
         'Sales Consolidation',
         'Investors Relations',
         'Abbott Shared Financial Services']

skills = ['Python, Mathematics, Predictive Methods, Deep Learning',
          'SQL (SSRS/Postgres), Report Building, Dashboarding & Analytics',
          'Concise Writing, Project Management',
          'Data Analysis & Forecasting, Project Management, Leadership',
          'Forecasting, Modelling',
          'Variance Analysis, Modelling',
          'Forecasting, Modelling, International Experience',
          'Data Management, Forecasting, Modelling',
          'Attention to Detail, Organization',
          'Attention to Detail, Project Management, Modelling'
         ]
  

p = figure(y_range=slots, plot_height=250,plot_width=800, x_range=(2013, 2019.5), title="Career Timeline - Hover for Info",
           toolbar_location=None,active_drag=None, active_scroll=None
          )

renderers = p.hbar_stack(layers, y='roles', height=.9,  source=ColumnDataSource(jobs), name=layers,
             color=['rgba(255,255,255,0)', 'rgba(199,92,220,1.0)', 'rgb(25,130,248)', 'grey', 'rgb(25,130,248)',
                    'rgb(25,130,248)', 'rgb(63,150,249)','rgba(101,171,250,1.0)', 'rgb(139,192,251)',
                    'rgb(177,212,252)', 'rgb(196,223,253)'],
             line_alpha=0,
            )
for i, r in enumerate(renderers[1:]):
    pos = r.name
    d = co[i]
    s = skills[i]
    hover = HoverTool(tooltips=[
        ("Role", "{} - {}".format(pos, d)),
        ("Skills", "{}".format(s))
    ], renderers=[r])
    p.add_tools(hover)

p.y_range.range_padding = 0
p.ygrid.grid_line_color = None
p.axis.minor_tick_line_color = None
p.yaxis.major_tick_line_color = None
p.xaxis.major_tick_line_color = 'grey'
p.xgrid.grid_line_color = 'grey'
p.outline_line_color = None
p.background_fill_color = '#2a2a2a'
p.border_fill_color = '#2a2a2a'

p.xaxis.axis_line_color = 'rgba(0,0,0,0)'
p.yaxis.axis_line_color = 'grey'
p.axis.major_label_text_color = 'white'
p.title.text_color = 'white'
p.title.align = 'center'

c = 2013.3
mtop = 1.33
ttop = 43.9

p.quad(top=[mtop+.2], bottom=[mtop-.97], left=[c-.15],
       right=[c+.9], color='#2a2a2a',
      line_color='grey', line_width=1, line_dash='solid')

key_employment = Label(x=c, y=ttop, x_units='data', y_units='screen',
                 text='Employment', render_mode='css', text_align='left',
                 text_color = 'white', text_font_size='8pt',
                 border_line_alpha=0, border_line_color = 'rgba(101,171,250,0)',
                 background_fill_color='rgba(101,171,250,0)')
p.add_layout(key_employment)

p.square(c-.065, mtop, size=6, color='rgb(25,130,248)')

key_edu = Label(x=c, y=ttop-15, x_units='data', y_units='screen',
                 text='Education', render_mode='css', text_align='left',
                 text_color = 'white',text_font_size='8pt',
                 border_line_alpha=0, border_line_color = 'rgba(199,92,220,0)',
                 background_fill_color='rgba(199,92,220,0)')
p.add_layout(key_edu)
p.square(c-.065, mtop-.38, size=6,  color='rgba(199,92,220,1)')


key_ent = Label(x=c, y=ttop-30, x_units='data', y_units='screen',
                 text='Entrepreneurship', render_mode='css', text_align='left',
                 text_color = 'white', text_font_size='8pt',
                 border_line_alpha=0, border_line_color = 'grey',
                 background_fill_alpha=0)
p.add_layout(key_ent)
p.square(c-.065, mtop-.76, size=6,  color='grey')

handle = show(p)
```
